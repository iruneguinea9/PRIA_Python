## EJERCICIO 2

Crea una funci칩n que dado un color en forma RGB lo transforme en su forma hexadecimal.

````python
# la funci칩n :)
def rgb_a_hexadecimal(rgb):
    # {:02x} especifica el formato de hexadecimal, a침ade 0 si necesita, y que sea de 2 caracteres
    # con format le especifico que ser치 rgb
    return '#{:02x}{:02x}{:02x}'.format(*rgb)

# una pruebita
color_rgb = (255, 165, 0)  
color_hexadecimal = rgb_a_hexadecimal(color_rgb)
print(f"El color RGB {color_rgb} en formato hexadecimal es: {color_hexadecimal}")
````

## EJERCICIO 3

Dibuja la funci칩n sin(x)/x de color #f06741, la funci칩n (2sin(x))/x de color #a4f041 y la funci칩n -sin(x)/x de color #8d41f0 donde x 갣 ```[-5, 5]``` en el mismo plot.

Crea el array x de tama침o 100 con el m칠todo .linspace() de numpy.

Haz que el tama침o de la figure sea 8 x 6.

````python
# el metodo linspace le dices los l칤mites y el n칰mero de elementos del array
x = np.linspace(-5 * np.pi, 5 * np.pi, 100)

# defino las funciones para dibujar
def f1(x):
    return np.sin(x) / x

def f2(x):
    return 2 * np.sin(x) / x

def f3(x):
    return -np.sin(x) / x

# el tama침o del gr치fico
plt.figure(figsize=(8, 6))

# sobre el array x, le aplica la funci칩n que corresponde, con su color y etiqueta
plt.plot(x, f1(x), color='#f06741', label='sin(x)/x')
plt.plot(x, f2(x), color='#a4f041', label='(2sin(x))/x')
plt.plot(x, f3(x), color='#8d41f0', label='-sin(x)/x')

# un titulo para el gr치fico
plt.title('sin(x)/x, (2sin(x))/x, y -sin(x)/x')
# nombres de columnas
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
````

## EJERCICIO 6

Dibuja el siguiente grafo con los nodos de color #ad80fa, las aristas de color #b1ddf5, el nombre de los nodos de color blanco y en negrita.
```
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
```

````python
# Creamos un grafo vac칤o
G = nx.Graph()

# A침adimos nodos al grafo
G.add_nodes_from("abcde")

# A침adimos aristas al grafo con pesos
G.add_weighted_edges_from([
    ("a", "b", 1),
    ("a", "c", 2),
    ("a", "d", 2),
    ("a", "e", 5),
    ("b", "e", 4),
    ("c", "d", 3),
    ("d", "e", 1)
])

# Definimos colores para los nodos y las aristas
nodos = ["#ad80fa"] * len(G.nodes())  # Color rojo claro
aristas = ["#b1ddf5"] * len(G.edges())  # Color azul claro

# Calculamos la posici칩n de los nodos usando spring_layout
pos = nx.spring_layout(G)

# Creamos la figura 
plt.figure(figsize=(10, 8))

# Dibujar nodos con color rojo claro
nx.draw_networkx_nodes(G, pos, node_color=nodos, node_size=500)

# Dibujar etiquetas de los nodos en blanco con fuente negrita
nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")

# Dibujar aristas con color azul claro y grosor 2
nx.draw_networkx_edges(G, pos, edge_color=aristas, width=2)

# Agregar t칤tulo al grafo
plt.title("Grafo ej 6")

plt.show()
````

Considera el grafo del ejercicio anterior. El tama침o de los nodos debe ser 350 por el grado del nodo. El tama침o de la figura global debe ser de 8 x 6.

````python

# Calcular tama침o de los nodos
tamanio_nodos = [350 * (G.degree(n) + 1) for n in G.nodes()]

# Como en el anterior
nodos = ["#ad80fa"] * len(G.nodes())
aristas = ["#b1ddf5"] * len(G.edges())


pos = nx.spring_layout(G)

plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(G, pos, node_color=nodos, node_size=tamanio_nodos)
nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")
nx.draw_networkx_edges(G, pos, edge_color=aristas, width=2)

plt.axis('off')
plt.title("Grafo 350")
plt.show()
````
