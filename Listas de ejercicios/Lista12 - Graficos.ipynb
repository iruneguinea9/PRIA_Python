Crea una función que dado un color en forma RGB lo transforme en su forma hexadecimal.

````python
def rgb_a_hexadecimal(rgb):
    return '#{:02x}{:02x}{:02x}'.format(*rgb)

color_rgb = (255, 165, 0)  
color_hexadecimal = rgb_a_hexadecimal(color_rgb)
print(f"El color RGB {color_rgb} en formato hexadecimal es: {color_hexadecimal}")
````

## EJERCICIO 3

Dibuja la función sin(x)/x de color #f06741, la función (2sin(x))/x de color #a4f041 y la función -sin(x)/x de color #8d41f0 donde x ∈ ```[-5π, 5π]``` en el mismo plot.

Crea el array x de tamaño 100 con el método .linspace() de numpy.

Haz que el tamaño de la figure sea 8 x 6.

````python
x = np.linspace(-5 * np.pi, 5 * np.pi, 100)

def f1(x):
    return np.sin(x) / x

def f2(x):
    return 2 * np.sin(x) / x

def f3(x):
    return -np.sin(x) / x

plt.figure(figsize=(8, 6))

plt.plot(x, f1(x), color='#f06741', label='sin(x)/x')
plt.plot(x, f2(x), color='#a4f041', label='(2sin(x))/x')
plt.plot(x, f3(x), color='#8d41f0', label='-sin(x)/x')

plt.title('sin(x)/x, (2sin(x))/x, y -sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
````

## EJERCICIO 4

Dibuja las 3 funciones del ejercicio anterior cada una en un subplot con el método .subplot().

Haz que el tamaño de la figure global sea 10 x 5.

````python

fig = plt.figure(figsize=(10, 5))

ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.plot(x, f1(x), color='#f06741', label='sin(x)/x')
ax1.set_title('sin(x)/x')

ax2.plot(x, f2(x), color='#a4f041', label='(2sin(x))/x')
ax2.set_title('(2sin(x))/x')

ax3.plot(x, f3(x), color='#8d41f0', label='-sin(x)/x')
ax3.set_title('-sin(x)/x')

for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
````

## EJERCICIO 5

Dibuja un diagrama de sectores con los siguientes datos y los siguientes colores. Añádele un título, etiquetas, sombra, una separación de 0.2 al sector más pequeño, haz que empiece en el ángulo 45 y muestra los porcentajes con un solo decimal.

options = ```["Suspenso", "Aprobado", "Notable", "Excelente"]```
count = ```[20, 10, 45, 25]```
colors = ```["#ff2667", "#ff7226", "#26a8ff", "#67ff26"]```

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'options': ["Suspenso", "Aprobado", "Notable", "Excelente"],
    'count': [20, 10, 45, 25]
})

data['percentage'] = data['count'].astype(float) / sum(data['count']) * 100

colors = ['#ff2667', '#ff7226', '#26a8ff', '#67ff26']

plt.figure(figsize=(10,8))
plt.pie(data['count'], labels=data['options'], colors=colors, startangle=45, autopct='%1.1f%%', shadow=True,  wedgeprops={'edgecolor':'black'},explode=[0, 0.2, 0, 0])

plt.title('Distribución de Calificaciones')

plt.legend(title='Opciones', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
````

## EJERCICIO 6

Dibuja el siguiente grafo con los nodos de color #ad80fa, las aristas de color #b1ddf5, el nombre de los nodos de color blanco y en negrita.
```
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
```

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])

node_colors = ["#ad80fa"] * len(G.nodes())
edge_colors = ["#b1ddf5"] * len(G.edges())

pos = nx.spring_layout(G)

plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=500)
nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2)

plt.axis('off')
plt.title("Grafo con Nodos Colorados")
plt.show()
````
