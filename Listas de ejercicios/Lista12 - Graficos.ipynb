## EJERCICIO 2

Crea una función que dado un color en forma RGB lo transforme en su forma hexadecimal.

````python
# la función :)
def rgb_a_hexadecimal(rgb):
    # {:02x} especifica el formato de hexadecimal, añade 0 si necesita, y que sea de 2 caracteres
    # con format le especifico que será rgb
    return '#{:02x}{:02x}{:02x}'.format(*rgb)

# una pruebita
color_rgb = (255, 165, 0)  
color_hexadecimal = rgb_a_hexadecimal(color_rgb)
print(f"El color RGB {color_rgb} en formato hexadecimal es: {color_hexadecimal}")
````

## EJERCICIO 3

Dibuja la función sin(x)/x de color #f06741, la función (2sin(x))/x de color #a4f041 y la función -sin(x)/x de color #8d41f0 donde x ∈ ```[-5π, 5π]``` en el mismo plot.

Crea el array x de tamaño 100 con el método .linspace() de numpy.

Haz que el tamaño de la figure sea 8 x 6.

````python
# el metodo linspace le dices los límites y el número de elementos del array
x = np.linspace(-5 * np.pi, 5 * np.pi, 100)

# defino las funciones para dibujar
def f1(x):
    return np.sin(x) / x

def f2(x):
    return 2 * np.sin(x) / x

def f3(x):
    return -np.sin(x) / x

# el tamaño del gráfico
plt.figure(figsize=(8, 6))

# sobre el array x, le aplica la función que corresponde, con su color y etiqueta
plt.plot(x, f1(x), color='#f06741', label='sin(x)/x')
plt.plot(x, f2(x), color='#a4f041', label='(2sin(x))/x')
plt.plot(x, f3(x), color='#8d41f0', label='-sin(x)/x')

# un titulo para el gráfico
plt.title('sin(x)/x, (2sin(x))/x, y -sin(x)/x')
# nombres de columnas
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
````

## EJERCICIO 6

Dibuja el siguiente grafo con los nodos de color #ad80fa, las aristas de color #b1ddf5, el nombre de los nodos de color blanco y en negrita.
```
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
```

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])

nodos = ["#ad80fa"] * len(G.nodes())
aristas = ["#b1ddf5"] * len(G.edges())

pos = nx.spring_layout(G)

plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color=nodos, node_size=500)
nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")
nx.draw_networkx_edges(G, pos, edge_color=aristas, width=2)

plt.title("Grafo con Nodos de colores")
plt.show()
````

## EJERCICIO 8

Considera el grafo del ejercicio anterior. El grosor de las aristas debe ir acorde al peso de cada arista.

````python

tamanio_nodos = [350 * (G.degree(n) + 1) for n in G.nodes()]

nodos = ["#ad80fa"] * len(G.nodes())
aristas = ["#b1ddf5"] * len(G.edges())

pos = nx.spring_layout(G)

plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G, pos, node_color=nodos, node_size=tamanio_nodos)

aristas_pesos = [edge[2]['weight'] * 2 for edge in G.edges(data=True)] # ojo con las multiplicaciones y los tipos...
nx.draw_networkx_edges(G, pos, edge_color=aristas, width=aristas_pesos)

nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")

plt.axis('off')
plt.title("Grafo con nodos de colores y aristas con pesos ")
plt.show()
````
